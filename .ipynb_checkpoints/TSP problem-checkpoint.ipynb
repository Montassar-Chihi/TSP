### First, we'll use a greedy algorithm 
* A greedy algorithm is any algorithm that follows the problem-solving heuristic of making the locally optimal choice at each stage

###### This is the algorithm we will use:



```python
shortest_paths = {}
for starting_town in G.nodes: # This algorithm was implimented to find shortest paths starting from all cities
    print("The town we started in is :",starting_town,"\n")
    current_town = starting_town # Intiliaze the current town
    total_distance = 0 # Intiliaze the total distance
    towns_visited = {} 
    towns_visited[current_town] = 0 # Adding the starting town to the visited ones to exclude it from the options later
    for i in range(len(G.nodes)-1):
        mini = 1000 # Intiliaze the minimal distance to something bigger than the biggest distance between towns
        for condidate_town in G.nodes:
            # Exclude the visited towns from the options
            if condidate_town in towns_visited: 
                continue
            # This is the distance  between the current town we are standing on and the condidate town that we are going
            # to compare with the last obtained minimum distance
            a_comparer = nx.dijkstra_path_length(G,current_town,condidate_town)
            if a_comparer < mini:
                # If the distance between the current town we are standing on and the condidate town is less than the last 
                # distance we got from comparing then that distance is the new minimum and the condidate town is now potentially
                # the next town we are going to 
                mini = a_comparer
                potential_town = condidate_town
        # After looping through all the towns now we are certain that the last potential town is the town with the min distance
        current_town = potential_town
        towns_visited[current_town] = mini
        # Calculating the total distance
        total_distance += mini
    # Adding the returning distance
    print("The last town we are visiting before returning to our initial town is :"
          , current_town ," with a distance of :", nx.dijkstra_path_length(G,current_town,starting_town),"\n")
    total_distance += nx.dijkstra_path_length(G,current_town,starting_town)
#     print(towns_visited,"\n")       # Path and the distance between nodes
    print(list(towns_visited.keys()),"\n")  # Path
    print("Total distance covered: ",total_distance,"\n")
    shortest_paths[str(total_distance)] = list(towns_visited.keys()) # Store the results
#     print("---------------------------------\n---------------------------------")
    break # But we will stop at Tunis
```

### We will now impliment the genetic algorithm
* In computer science and operations research, a genetic algorithm (GA) is a metaheuristic inspired by the process of natural selection that belongs to the larger class of evolutionary algorithms (EA). Genetic algorithms are commonly used to generate high-quality solutions to optimization and search problems by relying on biologically inspired operators such as mutation, crossover and selection.

###### This is the algorithm we will use:


```python
'''
The first generation have to be generated randomly and we will start from there.
Each time we will choose 2 parents according to a certain critirea in order to make a child_path that should provide a better
result than 'his parents' .
The critirea we will choose is the shortest distance. 
Each time we want to select parents, we will check if they meet this critirea using a FITNESS FUNCTION.
This function will return a probability of that parent (path) being selected.
We will then create a MATING POOL, a big list that represents the probabilities we calcualted using the fitnees function.
We will then apply CROSSOVER. We take some "chromosomes" (towns) from each parent and we make the new CHILD PATH.
Finally, we apply MUTATION, a way to ensure that there is enough "variety" to make a different and better child each generation.
'''

# Intiliaze the first generation randomly
def generate_population(population,population_size,towns,starting_town):
    for i in range(population_size):
        path = []
        path.append(starting_town)
        for t in random.sample(towns,len(towns)):
            path.append(t)
        path.append(starting_town)
        population.append(path)
    return population

# Calculate the distance covered in each path
def route(path):
    distance = sum(G[n][nbr]["weight"] for n, nbr in nx.utils.pairwise(path))
    return distance

# Calculate the longest (worst) distance in a population 
# This is used in order to normlize the fitness function using the max-min method
def max_distance(population):
    maxi = 0
    for path in population:
        if route(path) > maxi:
            maxi = route(path)
    return maxi

# Calculate the fitness of each path
def fitness(path,minimum_greedy,worst_distance):
    # First we calculate the distance of the path 
    distance = route(path)
    '''   
    The fitness score will depend on the distance covered by a path
    We will use a normalized score that scales from 1 and 0, 0 being the score of the longest path and 1 the score of the path 
    found by the greedy algorithm (since it is the threshhold we decided on). We use a normalized score so we can get rid of  
    the bad paths (the longest ones) before they get in the mating pool.
    If our fitness score is more than 1, then we know that the path we found is better than the greedy algorithm     
    '''
    score = 1-((distance-minimum_greedy)/(worst_distance-minimum_greedy))    
    return score

# Generate the mating pool, the big array that will make the new population
def generate_mating_pool(population,minimum_greedy):
    mating_pool = []
    # We will itterate through each path in our population 
    for path in population:
        # According to each path fitness score, we will assign a number of occurances to that path in the mating pool
        # meaning that if a path has a score of 0.1, it will be present in the mating pool 10 times
        number_of_repeatings = int(fitness(path,minimum_greedy,worst_distance) * 100 )
        # This is where we create the mating pool
        for i in range(number_of_repeatings):
            mating_pool.append(path)
    # We will then suffle the mating pool to ensure the randomness of selection
    random.shuffle(mating_pool)
    return mating_pool

# Apply Crossover
def crossover(mating_pool,starting_town,towns):
    child_path = []
    another_one = []
    # Pick 2 random parents (paths with higher scores will appear more in the mating pool and thus will be more likely to
    # to be picked)
    a = random.randint(0,len(mating_pool)-1) 
    b = random.randint(0,len(mating_pool)-1)
    parent1 = mating_pool[a]
    parent2 = mating_pool[b]
    child_path.append(starting_town) # Add the starting town first
    another_one.append(starting_town) # Add the starting town first
    # We will devide the parents into 5 sections and choose every time the section with the best fitness and add it to child
    for i in range(1,21,4):
        sub_parent1 = parent1[i:i+4]
        sub_parent2 = parent2[i:i+4]
        if (fitness(sub_parent1,minimum_greedy,worst_distance) <= fitness(sub_parent2,minimum_greedy,worst_distance)):
            child_path += sub_parent2
            another_one += sub_parent2
        else:
            child_path += sub_parent1
            another_one += sub_parent1
    # This method risks that child have towns that occurs more than once, we will take care of that 
    # But since we are randomly inserting the missing towns, we will try the adding missing towns in both first and 
    # second occurances and then we will select the one with better fitness, this won't cover all combinations but
    # it will add chances to get a better path.
    another_one.reverse() # reverse the list to make the last occurance the first one
    for town in child_path:
        if (child_path.count(town) != 1):
            for new_town in towns:
                if (new_town not in child_path):
                    child_path[child_path.index(town)] = new_town
                    another_one[another_one.index(town)] = new_town
                    break
    another_one.reverse() # undo the reverse to get you real path again
    another_one.append(starting_town) # Add the final distination town
    child_path.append(starting_town) # Add the final distination town
    # Decide which path to get based on the fitness score again
    if fitness(child_path,minimum_greedy,worst_distance) > fitness(another_one,minimum_greedy,worst_distance):
        return child_path
    else:
        return another_one

# Apply mutation
def mutation(child_path,mutation_rate):
    for i in child_path:
        # We will pick a random number, if it is less than the mutation_rate, we will take the town and swap it with 
        # another randomly
        if random.random() < mutation_rate:
            index_of_old_town = random.randint(1,len(child_path)-2)
            # In order to make sure that the random function does not chosoe the same town and swap it itself 
            # (thus,basically doing nothing), we will initilize the index with the same index and then keep
            # itterating until it is different
            index_of_new_town = index_of_old_town
            while(index_of_new_town == index_of_old_town):
                index_of_new_town = random.randint(1,len(child_path)-2)
            # This is where the swapping happens
            old_town = child_path[index_of_old_town]
            new_town = child_path[index_of_new_town]
            child_path[index_of_old_town] = new_town
            child_path[index_of_new_town] = old_town
    return child_path
```

```python
'''
These are the parameters we will be playing with in order to improve our algorithm performance
'''
starting_town = "Tunis"

## List of towns to visit
towns = [] 
# We will exclude the starting town from the towns that we are going to make the population with since it has to be the first 
# town in every path we create and we can't have it ordered randomly liek the others
for state in G.nodes:
    if state == starting_town :
        continue
    else:
        towns.append(state)
        
population_size = 250 # Number of paths we will use as a population

mutation_rate = 0.08 # Mutation Rate

population = [] # Initialize the population list

new_generation = [] # Initialize the new generation list

## Initialize a random shortest path for our loop to start
shoretest_path = []
shoretest_path.append(starting_town)
for t in random.sample(towns,len(towns)):
    shoretest_path.append(t)
shoretest_path.append(starting_town)

minimum_greedy = cost # The threshhold 

minimum_ga = 50000000 # The minimum that our genetic algorithm is producing

number_of_generations = 0 # Initialize the numebr of generations produced

old_path = [] # Initialize the old path list
compteur_of_improvment = 0 # Initialize the counter 
```

```python
''' TIME TO FIND THE SOLUTION !! '''

population = generate_population(population,population_size,towns,starting_town)
worst_distance = max_distance(population)
while((fitness(shoretest_path,minimum_greedy,worst_distance) <= 1) or (compteur_of_improvment < 10000) ):
    worst_distance = max_distance(population)
    mating_pool = generate_mating_pool(population,minimum_greedy)
    worst_distance = max_distance(population)
    for i in range(population_size):
        child_path = crossover(mating_pool,starting_town,towns)
        child_path = mutation(child_path,mutation_rate) 
        new_generation.append(child_path)
        for potential_shortest_path in new_generation:
#             print(potential_shortest_path)
            if route(potential_shortest_path) <= minimum_ga:
                minimum_ga = route(potential_shortest_path)
                shoretest_path = potential_shortest_path
    number_of_generations += 1
    # This counter is made to end the algorithm if the result does not improve after 10.000 itterations and is better than 
    # any older result 
    if (old_path == shoretest_path):
        compteur_of_improvment += 1
    else:
        compteur_of_improvment = 1
    old_path = shoretest_path
    population = new_generation
    new_generation = []
    print("numebr of generations we produced : ",number_of_generations)
    print("best minimum we got so far : ",minimum_ga)
    print("best path we got so far : ",shoretest_path)
```

## After trying a couple of parameters, we land on population size of 250 and mutation rate of 0.08.
## We notice that after 10000 generations the result we had won't improve , we can confirm that we have a better result than the greedy algorithm and this may be the optimal result, however the algorithm may need some tuning and may produce better result in less time
